```python
import matplotlib.pyplot as plt

models = ['Neural Network', 'LSTM', 'Grid Search CV', 'Decision Tree', 'Random Forest', 'Linear Regression', 'SDG Regressor','Gradient Booster']
rmse_values = [621.7498076752895, 51.91550456994024, 61.22818692829516, 71.36868555444056, 47.02730983904757, 42.24100052497699, 61.22818692829516,61.22818692829516]

# Plotting the RMSE values
plt.figure(figsize=(10, 6))
plt.bar(models, rmse_values)
plt.title('RMSE values for Different Models')
plt.xlabel('Models')
plt.ylabel('RMSE')
plt.xticks(rotation=45)

# Adding RMSE values on top of the bars
for i in range(len(models)):
    plt.text(i, rmse_values[i] + 5, round(rmse_values[i], 2), ha='center', fontsize=10)

plt.show()
```
